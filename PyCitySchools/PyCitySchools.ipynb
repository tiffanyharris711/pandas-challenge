```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#collect data values to store in District Summary dataframe
school_count = len(school_data_complete["school_name"].unique()) 
        #use unique method since every school repeats for every student on the merged data

student_count = len(school_data_complete["student_name"])

school_budget = school_data["budget"].sum() #used orig school dataset to avoid budget dups in merged data

avg_math_score = school_data_complete["math_score"].mean()

avg_reading_score = school_data_complete["reading_score"].mean()

math_passing_count = len(school_data_complete["math_score"].loc[school_data_complete["math_score"]>=70])
math_percent_pass = math_passing_count/student_count

reading_passing_count = len(school_data_complete["reading_score"].loc[school_data_complete["reading_score"]>=70])
reading_percent_pass = reading_passing_count/student_count

overall_passing_count = len(school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)])
overall_percent_pass = overall_passing_count/student_count
```

```python
#create District Summary dictionary
district_summary_data = {"Total Schools":[school_count], 
     "Total Students": [student_count], 
     "Total Budget": [school_budget], 
     "Average Math Score": [avg_math_score],
     "Average Reading Score": [avg_reading_score],
     "% Passing Math": [math_percent_pass],
     "% Passing Reading": [reading_percent_pass],
     "% Overall Passing": [overall_percent_pass]
    }
#create District Summary dataframe
district_summary_df = pd.DataFrame(data=district_summary_data)

#clean up formatting in District Summary dataframe
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["% Passing Math"] = (district_summary_df["% Passing Math"]*100).map("{:.2f}%".format)
district_summary_df["% Passing Reading"] = (district_summary_df["% Passing Reading"]*100).map("{:.2f}%".format)
district_summary_df["% Overall Passing"] = (district_summary_df["% Overall Passing"]*100).map("{:.2f}%".format)

#display the District Summary dataframe
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
summary_df = school_data.set_index(["school_name"])["type"]

school_data_groupBy = school_data_complete.groupby("school_name")
num_of_stu = school_data_complete["school_name"].value_counts()
budget = school_data_groupBy["budget"].mean()
per_stu_budget = budget/num_of_stu
read_avg = school_data_groupBy["reading_score"].mean()
math_avg = school_data_groupBy["math_score"].mean()
stu_pass_math= school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
per_pass_math = stu_pass_math/num_of_stu
stu_pass_read= school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
per_pass_read = stu_pass_read/num_of_stu
stu_pass_overall= school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")["Student ID"].count()
per_pass_overall = stu_pass_overall/num_of_stu
```

```python
school_summary_df = pd.DataFrame({
    "School Type": summary_df, 
    "Total Students": num_of_stu,
    "Total School Budget": budget,
    "Per Student Budget": per_stu_budget,
    "Average Math Score": math_avg,
    "Average Reading Score": read_avg,
    "Percent Passing Math": per_pass_math,
    "Percent Passing Reading": per_pass_read,
    "Percent Passing Overall": per_pass_overall
    })
school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
school_data_9th = school_data_complete.loc[school_data_complete["grade"]=="9th"].groupby("school_name", as_index=False)
school_data_10th = school_data_complete.loc[school_data_complete["grade"]=="10th"].groupby("school_name", as_index=False)
school_data_11th = school_data_complete.loc[school_data_complete["grade"]=="11th"].groupby("school_name", as_index=False)
school_data_12th = school_data_complete.loc[school_data_complete["grade"]=="12th"].groupby("school_name", as_index=False)

avg_school_data_9th = school_data_9th.mean()
avg_school_data_10th = school_data_10th.mean()
avg_school_data_11th = school_data_11th.mean()
avg_school_data_12th = school_data_12th.mean()
```

```python
avg_math_9th = avg_school_data_9th[["school_name","math_score"]].rename(columns={"math_score": "Avg Math Score 9th"})
avg_math_10th = avg_school_data_10th[["school_name","math_score"]].rename(columns={"math_score": "Avg Math Score 10th"})
avg_math_11th = avg_school_data_11th[["school_name","math_score"]].rename(columns={"math_score": "Avg Math Score 11th"})
avg_math_12th = avg_school_data_12th[["school_name","math_score"]].rename(columns={"math_score": "Avg Math Score 12th"})

math_data_complete = pd.merge(avg_math_9th, avg_math_10th, on=["school_name", "school_name"])
math_data_complete = pd.merge(math_data_complete, avg_math_11th, on=["school_name", "school_name"])
math_data_complete = pd.merge(math_data_complete, avg_math_12th, on=["school_name", "school_name"])

math_data_complete["Avg Math Score 9th"] = math_data_complete["Avg Math Score 9th"].map("{:.2f}".format)
math_data_complete["Avg Math Score 10th"] = math_data_complete["Avg Math Score 10th"].map("{:.2f}".format)
math_data_complete["Avg Math Score 11th"] = math_data_complete["Avg Math Score 11th"].map("{:.2f}".format)
math_data_complete["Avg Math Score 12th"] = math_data_complete["Avg Math Score 12th"].map("{:.2f}".format)

math_data_complete.rename(columns={"school_name": "School Name"})
```
